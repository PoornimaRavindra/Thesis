```python
NO = []
NO2 = []
NOx = []
NH3 = []
CO = []
SO2 = []
```

```python
for i in range(0,l):
    a = test_df.loc[start:end , ['NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2']]
#     a.head()
    print(a.head())
    NO.append(a['NO'].mean())
    NO2.append(a['NO2'].mean())
    NOx.append(a['NOx'].mean())
    NH3.append(a['NH3'].mean())
    CO.append(a['CO'].mean())
    SO2.append(a['SO2'].mean())
    start += 30
    end += 30
```

```python
NO
```

```python
datas = {'NO': NO,
        'NO2': NO2,
        'NOx': NOx,
        'NH3': NH3,
        'CO': CO,
        'SO2': SO2
        }

df = pd.DataFrame(datas, columns = ['NO', 'NO2','NOx','NH3','CO','SO2'])

print (df)
```

```python
from statsmodels.tsa.ar_model import AR
```

```python
def df_cluster(data):
    X = data
    train, test = X[0:-1], X[-1:]
    l = len(train) + 2
    model = AR(X)
    model_fit = model.fit()
    predictions = model_fit.predict(start=len(train), end=l, dynamic=True)
    return predictions
```

```python
no = df_cluster(df['NO'])
no2 = df_cluster(df['NO2'])
nox = df_cluster(df['NOx'])
nh3 = df_cluster(df['NH3'])
co = df_cluster(df['CO'])
so2 = df_cluster(df['SO2'])
```

```python
datas = {'NO': no,
        'NO2': no2,
        'NOx': nox,
        'NH3': nh3,
        'CO': co,
        'SO2': so2
        }

df = pd.DataFrame(datas, columns = ['NO', 'NO2','NOx','NH3','CO','SO2'])

print (df)
```
